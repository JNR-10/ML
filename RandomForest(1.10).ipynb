```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
digits.target
```

```python
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), digits.target, test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# ensemble is the term used when you are using multiple algorithms to predict the output
# here we are building multiple decision trees and coming with a vote to get the prediction
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# plotting a confusion matrix
# this gives is where the model failed

y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

# first parameter supplying the truth and then giving our predicted data
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# better visualisation using seaborn

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
